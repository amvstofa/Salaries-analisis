```python
# Ambil kolom numerik
numeric_cols = df.select_dtypes(include='number').columns.tolist()

cols = 2
rows = (len(numeric_cols) + 1) // cols

# Buat subplot grid
fig = make_subplots(rows=rows, cols=cols,
                    subplot_titles=numeric_cols,
                    vertical_spacing=0.08)

colors = px.colors.qualitative.Set2

# Loop semua kolom dan buat histogram
for idx, col in enumerate(numeric_cols):
    row = idx // cols + 1
    col_pos = idx % cols + 1

    fig.add_trace(go.Histogram(
        x=df[col],
        nbinsx=50,
        marker_color=colors[idx % len(colors)],
        name=col,
        showlegend=False
    ), row=row, col=col_pos)

fig.update_layout(
    height=550 * rows,
    width=900,
    title_text="Distribusi Kolom Numerik",
    bargap=0.05,
    template = "simple_white"
)

fig.show()
```

```python
salary_tredns = df.groupby('work_year')['salary_in_usd'].agg(['mean', 'median']).reset_index()

fig = px.line(salary_tredns,
              x="work_year",
              y=["mean", "median"],
              markers=True,
              color_discrete_sequence=px.colors.qualitative.Set2,
              title = "Salary trend 2020 - 2025",
              height=500,
              width=900)


fig.update_layout(
    template = "simple_white",
    yaxis_title="Salary in USD",
    xaxis_title="Work Year",
    hovermode="x unified")

fig.show()
```

**📌 Insight:**
- Salary Menunjukan tren kenaikan pada setiap tahunnya

```python
experience = df.groupby(["experience_level", "work_year"])["salary_in_usd"].mean().reset_index()

fig = px.line(experience,
             x="work_year",
             y="salary_in_usd",
             markers=True,
             color="experience_level",
             color_discrete_sequence=px.colors.qualitative.Set2,
             title = "Distribusi rata-rata Salary (USD) dari bedasarkan Experience Level")

fig.update_layout(
    template = "simple_white",
    yaxis_title="Salary in USD",
    xaxis_title="Work Year",
    hovermode="x unified")


fig.show()
```

```python
experience = df.groupby("experience_level")["remote_ratio"].value_counts().reset_index(name="count")

fig = px.bar(
    experience,
    x="experience_level",
    y="count",
    color="remote_ratio",
    title="Remote Ratio by Experience Level",
    barmode="group",
    color_discrete_sequence=px.colors.qualitative.Set2
)


fig.update_layout(
    template = "simple_white",
    yaxis_title="Count",
    xaxis_title="Experience Level",
)

fig.show()
```

**📌 Insight**
- Mayoritas dari experience yang ada dapat disimpulkan bahwa sistem kerja traditional masih menjadi opsi utama

```python
job = df.groupby('job_title')['salary_in_usd'].mean().reset_index()

top_job = job.nlargest(10, "salary_in_usd")
top_job["salary_text"] = top_job["salary_in_usd"].apply(lambda x: f"${x:,.2f}")

fig = px.bar(top_job,
             y="job_title",
             x="salary_in_usd",
             color="job_title",
             title="Rata-rata Gaji dalam USD untuk 10 Job Title Bergaji Tertinggi",
             color_discrete_sequence=px.colors.qualitative.Set2,
             text="salary_text")

fig.update_layout(
    template = "simple_white",
    yaxis_title="Job Title",
    xaxis_title="Salary in USD")

fig.show()
```

**📌 Insight:**
- Research Team Lead menduduki posisi teratas dengan gaji rata-rata sebesar $450,000, yang sangat tinggi dibandingkan dengan posisi lainnya.
- Dominasi Posisi tinggi dalam grafik yang ditunjukan menandakan bahwa, jabatan level senior atau eksekutif secara konsisten menerima gaji tertinggi.

```python
job = df.groupby('job_title')['salary_in_usd'].mean().reset_index()

below_salary_jobs = job.nsmallest(10, "salary_in_usd")
below_salary_jobs["salary_text"] = below_salary_jobs["salary_in_usd"].apply(lambda x: f"${x:,.2f}")

fig = px.bar(below_salary_jobs,
             y="job_title",
             x="salary_in_usd",
             color="job_title",
             title="Rata-rata Gaji dalam USD untuk 10 Job Title Bergaji Terendah",
             color_discrete_sequence=px.colors.qualitative.Set2,
             text="salary_text")

fig.update_layout(
    template = "simple_white",
    yaxis_title="Job Title",
    xaxis_title="Salary in USD")

fig.show()
```

**📌Insight**
- Posisi dengan gaji terendah berasal dari job, entry level,. Namun, ada anomali pada posisi yang biasanya bergaji tinggi (seperti AI Engineering Lead)

```python
country = df.groupby("company_location")["salary_in_usd"].mean().reset_index()

top_country = country.nlargest(10, "salary_in_usd")

fig = px.bar(top_country,
              x="company_location",
              y="salary_in_usd",
              color="company_location",
              color_discrete_sequence=px.colors.qualitative.Set2,
             title= "10 Negara dengan Rata-Rata Gaji Tertinggi di Bidang Data Science (USD)")

fig.update_layout(
    template = "simple_white",
    yaxis_title="Salary in USD",
    xaxis_title="Company Location")

fig.show()
```

**📌 Insight:**
- Negara-negara seperti Qatar, Congo, Venezuela dan United States menawarkan rata-rata gaji tertinggi dalam bidang data science secara global. Data ini telah dikonversi ke USD, sehingga mencerminkan nilai gaji yang adil untuk perbandingan lintas negara.

## 📌 **Conclusion**

Berdasarkan analisis tren salary dan berbagai variabel yang memengaruhinya, beberapa kesimpulan utama dapat ditarik sebagai berikut:

### 1. **Tren Kenaikan Gaji**
- Rata-rata gaji dalam bidang data menunjukkan **tren peningkatan dari tahun ke tahun**, meskipun ada beberapa fluktuasi pada kategori tertentu.
- Kenaikan ini konsisten pada hampir semua **tingkat pengalaman**, menandakan adanya perkembangan positif terhadap kompensasi profesional di bidang ini.

### 2. **Perbedaan Berdasarkan Tipe Employment**
- Tipe pekerjaan **Full-time** cenderung menunjukkan tren yang **stabil** dalam hal salary.
- Sebaliknya, **Contract** sempat mengalami lonjakan pada 2021, namun mengalami penurunan di tahun-tahun berikutnya.
- **Freelance** hanya memiliki data hingga 2024, namun cenderung menunjukkan jumlah yang terbatas.

### 3. **Pengaruh Ukuran Perusahaan**
- Perusahaan **berukuran Medium dan Large** menunjukkan **tren peningkatan gaji** yang signifikan tiap tahun.
- Sementara itu, perusahaan **berukuran Small** cenderung mengalami **penurunan rata-rata gaji**, yang mungkin menunjukkan keterbatasan anggaran atau kemampuan bersaing.

### 4. **Model Kerja: On-Site, Remote, dan Hybrid**
- Model kerja **On-Site** tetap menjadi pilihan utama banyak perusahaan, dengan tren gaji yang terus meningkat.
- Model **Remote** juga menunjukkan **pertumbuhan konsisten**, mencerminkan meningkatnya fleksibilitas dan adaptasi terhadap kerja jarak jauh.
- Sementara itu, model **Hybrid** justru menunjukkan tren **penurunan**, yang dapat mengindikasikan tantangan dalam implementasinya secara konsisten.

### 5. **Level Pengalaman dan Sistem Kerja**
- Sistem kerja tradisional (On-Site) masih dominan di semua tingkat pengalaman.
- Namun, **pekerja senior lebih cenderung mendapatkan opsi Remote**, menunjukkan adanya kepercayaan dan fleksibilitas lebih besar pada level tersebut.
- Model Hybrid belum stabil dan belum tersebar merata di seluruh tingkat pengalaman.

### 6. **Perbandingan Gaji Berdasarkan Job Title**
- **Posisi seperti Research Team Lead dan eksekutif senior lainnya** menempati posisi teratas dalam hal rata-rata gaji.
- Sebaliknya, posisi **entry-level** secara umum menerima gaji terendah.
- Ditemukan beberapa **anomali**, seperti beberapa peran teknis tinggi (misalnya AI Engineering Lead) yang memiliki rata-rata gaji lebih rendah dari ekspektasi.

### 7. **Gaji Berdasarkan Negara**
- Negara-negara seperti **Qatar, Congo, Venezuela, dan Amerika Serikat** menawarkan rata-rata gaji tertinggi dalam bidang data.
- Informasi ini dapat menjadi referensi penting dalam **pengambilan keputusan karier global**, baik untuk pekerja lokal maupun internasional
